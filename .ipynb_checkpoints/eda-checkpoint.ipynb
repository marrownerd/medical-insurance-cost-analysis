```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Настраиваем стиль и размер графиков
sns.set_style('whitegrid') # Стиль "белая сетка" для графиков
plt.rcParams['figure.figsize'] = (10, 6) # Устанавливаем размер графиков по умолчанию

print('Библиотеки импортированы, настройки применены!')
df = pd.read_csv('D:\projects\medical-insurance-cost-analysis\insurance.csv')
df.head()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data=df, x='smoker', y='charges')
plt.title('Стоимость страховки для курящих и некурящих', fontsize=16)
plt.xlabel('Курильщик', fontsize=12)
plt.ylabel('Стоимость (charges)', fontsize=12)
plt.show()
```

```python
df.groupby('smoker')['charges'].mean()
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(data=df, x='sex', y='charges')
plt.title('Стоимость страховки в зависимости от пола', fontsize=16)
plt.xlabel('Пол', fontsize=12)
plt.ylabel('Стоимость (charges)', fontsize=12)
plt.show()
```

```python
df.groupby('sex')['charges'].mean()
```

```python
plt.figure(figsize=(12, 7))
sns.boxplot(data=df, x='region', y='charges')
plt.title('Стоимость страховки по регионам', fontsize=16)
plt.xlabel('Регион', fontsize=12)
plt.ylabel('Стоимость (charges)', fontsize=12)
plt.show()
```

```python
df.groupby('region')['charges'].describe()
```
